### **Series**

* A Series is used to model one-dimensional data. 

* A Series is a one-dimensional labeled array capable of holding any data type. It consists of two main components: the index and the data. The index represents the labels associated with the data.

* it can store same and different type of data.
* it store data in sequentional order 

* Series ia a predefined class, having methods. We can use methods to perform operations on Series Values.

* We can Create a *Series* using *pd.Series()* and passing it a Python *list*.(arrays, range and dataframe)



#### **Creating a Series**
To create a Series, you can pass various data structures like lists, arrays, or dictionaries to the pd.Series() function.

```python
# Empty Series object 

# we need to create object to store Series pre-defined class
s = pd.Series() # 'S' i Series is capital because its a Class in Pandas 
print(s)
print(type(s))
```

```python
# We can created Series by using list, assigning a name 

m = [56,45,35,41,44,55]
s = pd.Series(m, name = 'marks') # name is parameter if you want to name your data 
print(s)
```



The series is one-dimensional. However, this looks like it is two-dimensional. The leftmost column is the index, which contains entries for the index. The index is not part of the values. The generic name for an index is an axis, and the values of the index—0, 1, 2, 3—are called axis labels. The data—145, 142, 38, and 13—is also called the values of the series. The two dimensional structure in pandas—a DataFrame—has two axes, one for the rows and another for the columns.


The rightmost column in the output contains the values of the series—145, 142, 38, and 13. In this case, they are integers (the console representation says dtype: int64, dtype meaning data type, and int64 meaning 64-bit integer), but in general, the values of a Series can hold strings, floats,booleans, or arbitrary Python objects.

#### **Accesssing Elements by Lable**

If you have defined labels for the Series, you can access elements using those labels.

```python
data = [10, 20, 30, 40, 50]
labels = ['A', 'B', 'C', 'D', 'E']
series = pd.Series(data, index=labels)

print(series['A'])  # Access the element with label 'A'
print(series['C'])  # Access the element with label 'C'
```

#### **Modifying Elements in a Series:**

You can modify elements in a Series using the index or label.

* We can Modify an element in a Series using the index and lables, uses the assignment operator '='.

```python
# run the previous code before running this block
series[1] = 25  # Modify the second element
print(series)
series['C'] = 35  # Modify the element with label 'C'
print(series)
```

### **Series - Attributes**

Notes: Variables/ objects stored within the class are called as Attributes

As *Series* is a class, it contains Attribute.

* The attributes  provide information about the Series data and its characteristics.

| Attribute | Description |Examples|
|-----------|-------------|---------|
|values|Returns the underlying data as Numpy array.|var_nam.values|
|index|Returns the index lable of series.|var_nam.index|
|dtype|Returns the data type of series.|var_nam.dtype|
|size|Returns the number of elements in series.|var_nam.size|
|shape|Returns a tuple representing the shape of the Series (number of rows, number of columns).|var_nam.shape|
|name|Returns the name of the Series.|var_nam.name|
|is_unique|Returns a Boolean indicating whether all the values in the Series are unique.|var_nam.is_unique|
|isnull()|Returns a Series of Booleans indicating which values are missing (NaN).|var_nam.isnull()|
|notnull()|Returns a Series of Booleans indicating which values are not missing (not NaN).|var_nam.notnull()|
|nunique()|Returns the number of unique values in the Series.|var_nam.nunique|




```python

data = [10, 20, 30, 40, 50]
index = ['A', 'B', 'C', 'D', 'E']
series = pd.Series(data, index=index, name='MySeries')

print("values")
print(series.values)

print("--- index ---")
print(series.index)
print("--- dtype ---")
print(series.dtype)
print(" --- size ---")
print(series.size)
print("--- shape ---")
print(series.shape)
print("--- name ---")
print(series.name)
print("--- is_unique ---")
print(series.is_unique)
print("--- isnull() ---")
print(series.isnull())
print("--- notnull() ---")
print(series.notnull())
print("--- nunique() ---")
print(series.nunique())
```

### **Series Methods**



Note: the functions whihc are defined within a class are called methods. 


| Method | Description |
|---|---|
| `at()` | Access a value at a particular index. |
| `iat()` | Access a value at a particular index by position. |
| `loc()` | Access a value or a range of values by label. |
| `iloc()` | Access a value or a range of values by position. |
| `head()` | Return the first few rows of the Series. |
| `tail()` | Return the last few rows of the Series. |
| `describe()` | Generate descriptive statistics for the Series. |
| `value_counts()` | Count the number of occurrences of each value in the Series. |
| `sort_values()` | Sort the Series by its values. |
| `to_numpy()` | Convert the Series to a NumPy array. |


For a more complete list of Series methods, please refer to the Pandas documentation: https://pandas.pydata.org/docs/reference/api/pandas.Series.html

```python
import pandas as pd
import numpy as np

marks = [56,45,35,41,60,57,56,56,np.nan]
s = pd.Series(marks)
```

```python
print(s) # this is used for series 

print("--head--")
print(s.head()) #this method returns first five values from the series 

print("--Tail--")
print(s.tail()) #this method returns last five values from the series 

print("--Sum--")
print(s.sum()) #it returns sum of all values 

print("--count--")
print(s.count()) #this method return number of non-nan/null values.

print("--mean--")
print(s.mean()) #this method returns mean of series values 

print("--describe--")
print(s.describe()) #it generate descriptive statistics for a Series or DataFrame, such as count, mean, standard deviation, minimum, maximum, etc.

print("--Unique--")
print(s.unique()) #it returns values from the series 

print("--nunique--")
print(s.nunique()) #it returns number of unique values 
```

## **Data Frames**


A DataFrame in Python is a two-dimensional data structure that is used to store and manipulate data. It is similar to a spreadsheet, but it is more powerful and flexible. DataFrames can be used to store data of any type, and they can be easily manipulated using the Pandas library.

Here are some of the key features of DataFrames:

* They are two-dimensional, meaning that they have rows and columns.
* They can store data of any type, including numbers, strings, dates, and objects.
* They can be easily manipulated using the Pandas library.
* They are a powerful tool for data analysis and visualization.

```python


df = pd.DataFrame({'data1': [1, 2, 3, 4, 5], 'data2': [6, 7, 8, 9, 10]})

print(df)
print(type(df))
```

we can create DataFrame in different ways like below.
